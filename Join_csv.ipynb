```python
import pandas as pd 
import glob
```

```python
path = "C:\\Users\Lenovo\Desktop\DataAugmentationForObjectDetection\\"
filenames = glob.glob(path + "\*.csv")
```

```python
annotation = pd.DataFrame()
```

```python

for file in filenames:
   gh=pd.read_csv(file)
   line=pd.DataFrame({'ID':gh['annotation/filename'],'x1':gh['annotation/object/bndbox/xmin'],'y1':gh['annotation/object/bndbox/ymax'],'x2':gh['annotation/object/bndbox/xmax'],'y2':gh['annotation/object/bndbox/ymin']})
   annotation=pd.concat([annotation,line],ignore_index=True)
```
